```python
from pathlib import Path
import os

# Force notebook to run from Phase-2 root (so relative paths work)
project_root = Path.cwd().parent
os.chdir(project_root)

print("CWD:", Path.cwd())
print("data exists:", Path("data").exists())
print("models exists:", Path("models").exists())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import xgboost as xgb

plt.rcParams["figure.figsize"] = (14, 5)
plt.rcParams["axes.grid"] = True

DATA_PATH = Path("data/processed/features_lagroll_24h.parquet")
MODEL_PATH = Path("models/xgb_24h.json")

TARGET_COL = "target_demand_mw_t_plus_24h"
FEATURE_COLS = [
    "hour_sin", "hour_cos",
    "dow_sin", "dow_cos",
    "is_weekend",
    "month_sin", "month_cos",
    "demand_lag_1",
    "demand_lag_24",
    "demand_lag_48",
    "demand_lag_168",
    "demand_roll_mean_24",
    "demand_roll_std_24",
    "demand_roll_mean_168",
    "demand_roll_std_168",
]

TRAIN_END = "2025-06-30 23:00:00+00:00"
VAL_END   = "2025-09-30 23:00:00+00:00"
```

```python
df = pd.read_parquet(DATA_PATH)
df["timestamp"] = pd.to_datetime(df["timestamp"], utc=True)
df = df.sort_values("timestamp").reset_index(drop=True)

# Safety: should already be clean
df = df.dropna(subset=[TARGET_COL] + FEATURE_COLS).reset_index(drop=True)

train_end_ts = pd.to_datetime(TRAIN_END, utc=True)
val_end_ts = pd.to_datetime(VAL_END, utc=True)

train_df = df[df["timestamp"] <= train_end_ts].copy()
val_df = df[(df["timestamp"] > train_end_ts) & (df["timestamp"] <= val_end_ts)].copy()
test_df = df[df["timestamp"] > val_end_ts].copy()

print("Rows:", len(df))
print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))
print("Range:", df["timestamp"].min(), "→", df["timestamp"].max())
```

```python
booster = xgb.Booster()
booster.load_model(str(MODEL_PATH))

def predict(df_part: pd.DataFrame) -> np.ndarray:
    dmat = xgb.DMatrix(df_part[FEATURE_COLS].to_numpy(), feature_names=FEATURE_COLS)
    return booster.predict(dmat)

val_df = val_df.copy()
test_df = test_df.copy()

val_df["y_true"] = val_df[TARGET_COL].to_numpy()
val_df["y_pred"] = predict(val_df)
val_df["resid"] = val_df["y_true"] - val_df["y_pred"]

test_df["y_true"] = test_df[TARGET_COL].to_numpy()
test_df["y_pred"] = predict(test_df)
test_df["resid"] = test_df["y_true"] - test_df["y_pred"]

def mae(y_true, y_pred):
    return float(np.mean(np.abs(y_true - y_pred)))

def rmse(y_true, y_pred):
    return float(np.sqrt(np.mean((y_true - y_pred) ** 2)))

print("VAL MAE:", mae(val_df["y_true"].values, val_df["y_pred"].values))
print("VAL RMSE:", rmse(val_df["y_true"].values, val_df["y_pred"].values))
print("TEST MAE:", mae(test_df["y_true"].values, test_df["y_pred"].values))
print("TEST RMSE:", rmse(test_df["y_true"].values, test_df["y_pred"].values))
```

```python
start = pd.Timestamp("2025-08-01 00:00:00+00:00")
end   = pd.Timestamp("2025-08-14 23:00:00+00:00")

slice_df = val_df[(val_df["timestamp"] >= start) & (val_df["timestamp"] <= end)].copy()

plt.figure()
plt.plot(slice_df["timestamp"], slice_df["y_true"], label="Actual")
plt.plot(slice_df["timestamp"], slice_df["y_pred"], label="Predicted")
plt.title("Demand Forecast (24h ahead) — Actual vs Predicted (Aug 2025 slice)")
plt.xlabel("Timestamp (UTC)")
plt.ylabel("MW")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.hist(val_df["resid"], bins=60, alpha=0.6, label="Validation")
plt.hist(test_df["resid"], bins=60, alpha=0.6, label="Test")
plt.title("Residual Distribution (y_true - y_pred)")
plt.xlabel("MW error")
plt.ylabel("Count")
plt.legend()
plt.show()

print("VAL resid mean:", float(val_df["resid"].mean()))
print("TEST resid mean:", float(test_df["resid"].mean()))
```

```python
naive_path = Path("reports/rolling_backtest_naive_24h.csv")
ridge_path = Path("reports/rolling_backtest_ridge_lagroll_24h.csv")
xgb_path = Path("reports/rolling_backtest_xgb_24h.csv")

naive_bt = pd.read_csv(naive_path, parse_dates=["test_start"])
ridge_bt = pd.read_csv(ridge_path, parse_dates=["test_start"])
xgb_bt = pd.read_csv(xgb_path, parse_dates=["test_start"])

naive_bt["month"] = naive_bt["test_start"].dt.strftime("%Y-%m")
ridge_bt["month"] = ridge_bt["test_start"].dt.strftime("%Y-%m")
xgb_bt["month"] = xgb_bt["test_start"].dt.strftime("%Y-%m")

plt.figure(figsize=(14, 5))
plt.plot(naive_bt["month"], naive_bt["mae"], marker="o", label="Naive (t-24)")
plt.plot(ridge_bt["month"], ridge_bt["mae"], marker="o", label="Ridge (lag+rolling)")
plt.plot(xgb_bt["month"], xgb_bt["mae"], marker="o", label="XGBoost (lag+rolling)")
plt.xticks(rotation=45)
plt.title("Rolling Backtest MAE by Month (24h ahead)")
plt.xlabel("Month")
plt.ylabel("MAE (MW)")
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Diagnostic Interpretation

# The naive seasonal baseline performs surprisingly well, confirming strong daily/weekly seasonality in NYISO demand.

# Ridge regression improves stability in shoulder months but struggles during high-volatility summer periods.

# XGBoost achieves the lowest median MAE across rolling backtests, particularly improving performance during transitional months.

# All models exhibit elevated error during extreme summer demand, indicating regime-driven behavior and limits of historical lag features alone.
```

```python
out_path = Path("reports/phase2_diagnostics_val_test_predictions.csv")
out_path.parent.mkdir(parents=True, exist_ok=True)

preds_export = pd.concat([
    val_df[["timestamp", "y_true", "y_pred", "resid"]].assign(split="val"),
    test_df[["timestamp", "y_true", "y_pred", "resid"]].assign(split="test"),
], ignore_index=True)

preds_export.to_csv(out_path, index=False)
print("Saved:", out_path)
preds_export.head()
```
