### 3) Sanity checks (trust but verify)

Before analyzing anything, we confirm the artifacts are valid:

- required columns exist
- timestamps are sorted (monotonic increasing)
- there are no missing values in target/prediction columns

These checks prevent silent issues from contaminating diagnostics.


```python
assert TS_COL in df.columns
assert PRICE_COL in df.columns
assert TARGET_COL in df.columns

assert TS_COL in preds.columns
assert "y_true" in preds.columns
assert "y_pred" in preds.columns

assert df[TS_COL].is_monotonic_increasing
assert preds[TS_COL].is_monotonic_increasing

df[[PRICE_COL, TARGET_COL]].isna().sum(), preds[["y_true", "y_pred"]].isna().sum()
```

### 4) Define evaluation metrics (MAE, RMSE)

We define the same metrics used in the Phase 2B scripts:

- **MAE**: typical hourly error (robust to spikes)
- **RMSE**: penalizes large errors (spike-sensitive)

We also define a small helper that summarizes metrics in a consistent format.

```python
def mae(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    return float(np.mean(np.abs(y_true - y_pred)))

def rmse(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    return float(math.sqrt(np.mean((y_true - y_pred) ** 2)))

def summarize_metrics(name, y_true, y_pred):
    return {
        "model": name,
        "rows": int(len(y_true)),
        "mae": mae(y_true, y_pred),
        "rmse": rmse(y_true, y_pred),
    }
```

### 5) Align timestamps and reconstruct the test set (timezone-safe)

This cell resolves a common real-world data issue:  
**timestamp dtype mismatches** between saved artifacts.

Specifically, it:

1. Forces both datasets (`df` and `preds`) to use **UTC timezone-aware timestamps**
2. Reconstructs the test dataset by aligning:
   - the timestamps used during XGBoost evaluation
   - the corresponding true price at time *t*
   - the true next-hour target price *(t+1)*
3. Performs sanity checks to ensure:
   - no missing prices
   - no missing targets
   - no silent merge failures

This step guarantees that all baseline and diagnostic comparisons are made on
the **exact same test timestamps** used by the trained model.


```python
# --- Fix timestamp dtype mismatch (naive vs tz-aware) ---
df[TS_COL] = pd.to_datetime(df[TS_COL], utc=True)
preds[TS_COL] = pd.to_datetime(preds[TS_COL], utc=True)

# Merge true price at time t and target at t+1 for the test timestamps
test_ts = preds[[TS_COL]].copy()

test = test_ts.merge(
    df[[TS_COL, PRICE_COL, TARGET_COL]],
    on=TS_COL,
    how="left",
    validate="one_to_one",
)

# Sanity checks
missing_price = test[PRICE_COL].isna().sum()
missing_target = test[TARGET_COL].isna().sum()
missing_any = test.isna().any(axis=1).sum()

print("Missing price rows:", missing_price)
print("Missing target rows:", missing_target)
print("Rows with any missing:", missing_any)
print("Test rows:", len(test))

# If you see any missing rows, we'll handle them cleanly (should usually be 0)
```

### 6) Rebuild baselines on the *same timestamps* as the XGBoost test set

To compare fairly, baselines must be evaluated on the **exact same test period** as the saved XGBoost predictions.

This cell:

1. Aligns timestamps between dataframes (timezone-safe)
2. Creates a `test` dataframe containing:
   - `price(t)` (current price)
   - `target_price(t+1)` (next-hour true price)
3. Builds two baseline predictors:
   - **Persistence**: `price(t+1) = price(t)`
   - **24h rolling mean** (causal): mean of the last 24 hours of price

This ensures a true apples-to-apples comparison with XGBoost.


```python
# Ensure timestamps are aligned (safe to run again)
df[TS_COL] = pd.to_datetime(df[TS_COL], utc=True)
preds[TS_COL] = pd.to_datetime(preds[TS_COL], utc=True)

# Recreate "test" cleanly from scratch
test_ts = preds[[TS_COL]].copy()

test = test_ts.merge(
    df[[TS_COL, PRICE_COL, TARGET_COL]],
    on=TS_COL,
    how="left",
    validate="one_to_one",
)

# Drop any rare unmatched timestamps (should be 0 usually)
test = test.dropna(subset=[PRICE_COL, TARGET_COL]).reset_index(drop=True)

# --- Baseline 1: Persistence ---
test["pred_persistence"] = test[PRICE_COL]

# --- Baseline 2: 24h rolling mean (causal) ---
df = df.sort_values(TS_COL).reset_index(drop=True)
df["pred_rollmean_24h"] = df[PRICE_COL].rolling(window=24, min_periods=24).mean()

test = test.merge(
    df[[TS_COL, "pred_rollmean_24h"]],
    on=TS_COL,
    how="left",
    validate="one_to_one",
)

# Quick confirmation
print("Columns in test:", list(test.columns))
print("Missing pred_persistence:", test["pred_persistence"].isna().sum())
print("Missing pred_rollmean_24h:", test["pred_rollmean_24h"].isna().sum())
print("Rows in test:", len(test))
```

### 7) Compare model performance (XGBoost vs baselines)

This cell computes **MAE** and **RMSE** for:

- XGBoost (from saved predictions)
- Persistence baseline
- Rolling mean baseline

The output table provides a clear benchmark:  
**XGBoost must beat persistence to be considered valuable.**


```python
rows = []

# XGB (from preds file)
rows.append(
    summarize_metrics("XGBoost", preds["y_true"], preds["y_pred"])
)

# Persistence
rows.append(
    summarize_metrics("Persistence", test[TARGET_COL], test["pred_persistence"])
)

# Rolling mean (drop NaNs that exist only if early window leaks into test)
roll_sub = test.dropna(subset=["pred_rollmean_24h"])
rows.append(
    summarize_metrics("RollMean24h", roll_sub[TARGET_COL], roll_sub["pred_rollmean_24h"])
)

metrics_df = pd.DataFrame(rows).sort_values("mae")
metrics_df
```

### 8) Quantify improvement vs the strongest baseline (persistence)

Electricity price is highly autocorrelated at short horizons, so the persistence baseline is a high bar.

This cell computes the percentage improvement of XGBoost vs persistence:

- MAE improvement (%)
- RMSE improvement (%)

This is the headline comparison used in the Phase 2B write-up.

```python
m_xgb = metrics_df.loc[metrics_df["model"] == "XGBoost"].iloc[0]
m_persist = metrics_df.loc[metrics_df["model"] == "Persistence"].iloc[0]

mae_improve = (m_persist["mae"] - m_xgb["mae"]) / m_persist["mae"] * 100
rmse_improve = (m_persist["rmse"] - m_xgb["rmse"]) / m_persist["rmse"] * 100

mae_improve, rmse_improve
```

### 9) Create residuals (error) for diagnostics

We compute:

- `err = y_pred - y_true` (signed error)
- `abs_err = |err|` (magnitude of error)

These are the foundation for:
- error distributions
- regime analysis (normal vs spikes)
- hour-of-day analysis

```python
diag = preds.copy()
diag["err"] = diag["y_pred"] - diag["y_true"]
diag["abs_err"] = diag["err"].abs()

diag.describe()[["y_true", "y_pred", "err", "abs_err"]]
```

### 12) Price regimes: normal vs high vs spike conditions

Electricity prices are heavy-tailed. Performance must be interpreted by regime.

We compute thresholds using the true price distribution on the test set:

- **p95** = high-price threshold
- **p99** = spike threshold

Then we compute MAE/RMSE separately for:
- Normal ( p95)
- High (p95p99)
- Spike (> p99)

This reveals where the model is reliable vs where it degrades.

```python
p95 = float(diag["y_true"].quantile(0.95))
p99 = float(diag["y_true"].quantile(0.99))

p95, p99
```

### 13) Aggregate performance by price regime (normal, high, spike)

Electricity prices are heavy-tailed, so average metrics alone can be misleading.

This cell defines a helper function that computes:
- MAE
- RMSE
- median absolute error
- row counts

We then split the test data into three **price regimes** based on true price quantiles:

- **Normal ( p95)**  typical operating conditions
- **High (p95p99)**  stressed but common high-price periods
- **Spike (> p99)**  extreme scarcity events

Aggregating metrics by regime allows us to:
- understand where the model is reliable
- identify where error growth is expected
- make honest claims about model limitations


```python
def regime_metrics(df_sub, name):
    return {
        "regime": name,
        "rows": int(len(df_sub)),
        "mae": mae(df_sub["y_true"], df_sub["y_pred"]),
        "rmse": rmse(df_sub["y_true"], df_sub["y_pred"]),
        "median_abs_err": float(df_sub["abs_err"].median()),
    }

normal = diag[diag["y_true"] <= p95]
high = diag[(diag["y_true"] > p95) & (diag["y_true"] <= p99)]
spike = diag[diag["y_true"] > p99]

regimes = pd.DataFrame([
    regime_metrics(normal, "Normal (<= p95)"),
    regime_metrics(high, "High (p95p99)"),
    regime_metrics(spike, "Spike (> p99)"),
])

regimes
```

### 14) Visualize performance by regime

We plot MAE and RMSE across the three regimes.

Expected behavior:
- strong performance in normal regime
- higher error in high regime
- very large error in spike regime (scarcity events)

If a model looks too good on spikes, that can be a leakage red flag.


```python
plt.bar(regimes["regime"], regimes["mae"])
plt.title("MAE by Price Regime")
plt.ylabel("MAE ($/MWh)")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()

plt.bar(regimes["regime"], regimes["rmse"])
plt.title("RMSE by Price Regime")
plt.ylabel("RMSE ($/MWh)")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()
```

### 15) Hour-of-day error analysis (UTC)

Electricity markets often show time-of-day structure due to:

- daily demand ramps
- peak vs off-peak patterns
- typical congestion periods

We compute MAE and RMSE by hour (UTC) to identify:

- when the model performs best
- when volatility makes forecasting harder

This is especially useful for operational planning and risk awareness.

```python
diag["hour_utc"] = pd.to_datetime(diag[TS_COL]).dt.hour

hour_metrics = (
    diag.groupby("hour_utc")
        .agg(
            rows=("y_true", "size"),
            mae=("abs_err", "mean"),
            rmse=("err", lambda s: float(math.sqrt(np.mean(np.asarray(s, dtype=float) ** 2)))),
        )
        .reset_index()
)

hour_metrics.head()

plt.plot(hour_metrics["hour_utc"], hour_metrics["mae"], marker="o")
plt.title("MAE by Hour (UTC)")
plt.xlabel("Hour (UTC)")
plt.ylabel("MAE ($/MWh)")
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()

plt.plot(hour_metrics["hour_utc"], hour_metrics["rmse"], marker="o")
plt.title("RMSE by Hour (UTC)")
plt.xlabel("Hour (UTC)")
plt.ylabel("RMSE ($/MWh)")
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

### 17) Spike honesty check: does persistence beat XGBoost on extreme events?

This optional step compares XGBoost vs persistence on the same timestamps.

Why:
- persistence is often hard to beat in sudden shocks
- this check ensures our model improvements are real and not just averages
- it gives an honest view of performance under stress (spikes)


```python
# Build a persistence prediction for the test timestamps (same timestamps)
# price(t) is available in `test` (merged from df using diag timestamps)
compare = diag[[TS_COL, "y_true", "y_pred"]].merge(
    test[[TS_COL, "pred_persistence"]],
    on=TS_COL,
    how="left",
    validate="one_to_one"
)

compare["abs_err_xgb"] = (compare["y_pred"] - compare["y_true"]).abs()
compare["abs_err_persist"] = (compare["pred_persistence"] - compare["y_true"]).abs()

worst_cmp = compare.sort_values("abs_err_xgb", ascending=False).head(20)
worst_cmp[[TS_COL, "y_true", "y_pred", "pred_persistence", "abs_err_xgb", "abs_err_persist"]]
```
